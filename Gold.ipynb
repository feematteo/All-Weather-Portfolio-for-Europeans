```python
import yfinance as yahooFinance
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = "2000-09-01"
end = '2023-11-03'
GOLD = yahooFinance.download('GC=F',start,end)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(GOLD.index, GOLD['Adj Close'], label='Gold price', color='black')

plt.title('Gold Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas_datareader as pdr
import datetime
```

```python
start_date = datetime.datetime(2000, 1, 1)
end_date = datetime.datetime(2022, 12, 31)

# Fetch US CPI data from FRED
inflation_data = pdr.get_data_fred('CPIAUCNS', start_date, end_date)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(inflation_data.index, inflation_data, label='Gold price', color='black')

plt.title('Gold Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
